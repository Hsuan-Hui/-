```python
from tkinter import *
import csv
import tkinter.messagebox
import random

class Main:
    current_wrds =''
    ans_wrds = ''
    def __init__(self):
        window = Tk() # Create a window
        window.title("英英每單字") # Set title
        window.geometry('300x300')
        window.resizable(0,0)
        window['bg']='#F5F5F5'
        
        frame1=Frame(window)
        frame2=Frame(window)
        frame3=Frame(window)
        frame4=Frame(window)
        
        self.dwd_text=Text(frame1,bg='#FDF5E6',height=7,width=30)
        self.dwd_text.pack(side=LEFT,fill=BOTH)
        frame1.pack()
        
        Button(frame2,text='文章選讀',bg="#FFCCCC",command=self.Article,height=2,width=9).pack(side=LEFT,fill=BOTH) #Article function
        frame2.pack()
        
        Button(frame3,text='我的單字庫',bg='#CCEEFF',command=self.Wordbank,height=2,width=9).pack(side=LEFT,fill=BOTH)     
        frame3.pack()
        
        Button(frame4,text='離開',bg='#CCEEFF',command=window.destroy,height=2,width=9).pack(side=LEFT,fill=BOTH)     
        frame4.pack()
        
        again='y'
        if again=='y':
            self.daily_wd()
            
        window.mainloop()
        
    def daily_wd(self):
        wdcsvfile=open('英文單字.csv',encoding='utf-8')
        wdreader=csv.reader(wdcsvfile)
        wrdbook={}
        for row in wdreader:
            wrdbook[row[0]]=row[1] + "\n" + row[2]

        voc_lst=list(wrdbook.keys())
        def_lst=list(wrdbook.values())

        dvoc=random.choice(voc_lst)
        ddef=wrdbook[dvoc]
        dwd=dvoc+'\n'+ddef
        self.dwd_text.insert(END,dwd) 
        wdcsvfile.close()
        
    def Wordbank(self):
        window = Tk() 
        window.title("我的單字庫") 
        window.geometry('300x300')
        window['bg']='#F5F5F5'
        window.resizable(0,0)
    
        Button(window, text ="新增單字",bg="#FFDAB9", command=self.enter_words,height=2,width=9).place(x=115,y=50)
        Button(window, text = "搜尋單字",bg="#FFDAB9", command=self.srh_words,height=2,width=9).place(x=115,y=100)
        Button(window, text = "離開",bg="#33cc99",command=window.destroy,height=2,width=11).place(x=108,y=150)

        window.mainloop()
    def enter_words(self):
                
        window = Tk() 
        window.title("新增單字") 
        window.geometry('300x300')
        window['bg']='#F5F5F5'
        window.resizable(0,0)
            
        Label(window, text='單字：').place(x=50,y=50)
        self.voc_entry=Entry(window,width=20)
        self.voc_entry.place(x=90,y=50)
        Label(window, text='解釋：').place(x=50,y=100)
        self.def_entry=Entry(window,width=20)
        self.def_entry.place(x=90,y=100)
        Label(window, text='例句：').place(x=50,y=150)
        self.ex_entry=Entry(window,width=20)
        self.ex_entry.place(x=90,y=150)
        Button(window, text='加入',bg="#FFDAB9", command=self.add_words).place(x=170,y=250)
        Button(window, text='離開',bg="#33cc99", command=window.destroy).place(x=100,y=250)
      

        window.mainloop()
        
    def add_words(self):
        with open('MyWordbank.csv','a',encoding='utf-8') as wdFile:
            wdwriter=csv.writer(wdFile)
            nvoc=str(self.voc_entry.get())
            ndef=str(self.def_entry.get())
            nex=str(self.ex_entry.get())
            wdwriter.writerow([nvoc,ndef,nex])
        #print([nvoc,ndef,nex])
        
        isOK=tkinter.messagebox.showinfo(title='Success',message='Success to add!')
        print(isOK)
        
    def srh_words(self):
        window = Tk() 
        window.title("搜尋單字") 
        window.geometry('300x300')
        window['bg']='#F5F5F5'
        window.resizable(0,0)
        
        frame1=Frame(window)
        frame2=Frame(window)
        frame3=Frame(window)
        
        
        Label(frame1, text='找單字：').pack(side='left')
        self.fd_entry=Entry(frame1,width=15)
        self.fd_entry.pack(side='left')
        frame1.pack()
        
        b2=Button(frame2, text='搜尋',bg="#FFDAB9", command=self.find_words)
        b1=Button(frame2, text='離開',bg="#33cc99", command=window.destroy)
        b1.pack(side='right')
        b2.pack(side='right')
        frame2.pack()
        
        self.fd_text=Text(frame3)
        self.fd_text.pack(side=LEFT,fill=Y)
        frame3.pack()
        
        window.mainloop()
        
    def find_words(self): 
        fd_words=str(self.fd_entry.get())
        with open('MyWordbank.csv','r',encoding='utf-8-sig') as fdfile:
            fdreader=csv.reader(fdfile)
            wrddbook={}
            try:
                for row in fdreader:
                    wrddbook[row[0]]=row[1] + "\n" + row[2]
                self.ans_wrds = wrddbook[fd_words]
                self.fd_text.insert(END,self.ans_wrds+"\n") 
            except:
                self.fd_text.insert(END,"Not found.\n")
                print(wrddbook)
                print(fd_words)
        
                
    def Article(self):
        window = Tk() 
        window.title("英英每單字") 
        window.geometry('300x300')
        window['bg']='#F5F5F5'
        window.resizable(0,0)

        Button(window,text='Business',bg="#FFDAB9",command=self.ba,height=2,width=9).place(x=115,y=50) 
        Button(window,text='Science',bg="#FFDAB9",command=self.sa,height=2,width=9).place(x=115,y=100) 
        Button(window,text='Life & Health',bg="#FFDAB9",command=self.ha,height=2,width=11).place(x=108,y=150)

        window.mainloop()

    def ba(self):
        window = Tk()
        window.title("英英每單字")
        window.geometry('300x300')
        window.resizable(0,0)
        Button(window,text='b01',command=lambda:self.name('c01.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b02',command=lambda:self.name('c02.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b03',command=lambda:self.name('c03.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b04',command=lambda:self.name('c04.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b05',command=lambda:self.name('c05.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b06',command=lambda:self.name('c06.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b07',command=lambda:self.name('c07.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b08',command=lambda:self.name('c08.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b09',command=lambda:self.name('c09.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='b10',command=lambda:self.name('c10.txt'),bg="#FFDAB9",width=8).pack()
        window.mainloop()

    def sa(self):
        window = Tk()
        window.title("英英每單字")
        window.geometry('300x300')
        window.resizable(0,0)
        Button(window,text='s01',command=lambda:self.name('s01.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s02',command=lambda:self.name('s02.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s03',command=lambda:self.name('s03.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s04',command=lambda:self.name('s04.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s05',command=lambda:self.name('s05.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s06',command=lambda:self.name('s06.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s07',command=lambda:self.name('s07.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s08',command=lambda:self.name('s08.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s09',command=lambda:self.name('s09.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='s10',command=lambda:self.name('s10.txt'),bg="#FFDAB9",width=8).pack()
        window.mainloop()
        
    def ha(self):
        window = Tk()
        window.title("英英每單字")
        window.geometry('300x300')
        window.resizable(0,0)
        Button(window,text='h01',command=lambda:self.name('h01.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h02',command=lambda:self.name('h02.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h03',command=lambda:self.name('h03.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h04',command=lambda:self.name('h04.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h05',command=lambda:self.name('h05.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h06',command=lambda:self.name('h06.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h07',command=lambda:self.name('h07.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h08',command=lambda:self.name('h08.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h09',command=lambda:self.name('h09.txt'),bg="#FFDAB9",width=8).pack()
        Button(window,text='h10',command=lambda:self.name('h10.txt'),bg="#FFDAB9",width=8).pack()
        window.mainloop()

    def name(self,filename):
        infile=open(filename,'r',encoding='utf-8')
        content=infile.read()
        infile.close() 
        window = Tk() 
        window.title("英英每單字")
        window.geometry('400x550')
        Button(window,text='查單字',command=self.wrds,bg='#F5FFFA',width=8).pack(side=TOP)
        window.resizable(0,0)
        scrollbar = Scrollbar(window)  
        scrollbar.pack(side=RIGHT,fill=Y)
        text = Text(window)
        text.pack(side=LEFT,fill=Y)
        text.config(yscrollcommand=scrollbar.set)
        scrollbar.config(command=text.yview)
        text.insert(END,content)
        window.mainloop()
        
    def wrds(self):
        window = Tk()
        window.title("英英每單字")
        window.geometry('400x400')
        window.resizable(0,0)
        frame1=Frame(window)
        frame2=Frame(window)
        frame3=Frame(window)
        
        Label(frame1,text='請輸入欲查詢的單字：').pack(side='left')
        self.entry = Entry(frame1,width=15)
        self.entry.pack(side='left')
        frame1.pack()
        
        b3= Button(frame2,text='收藏',bg="#FFDAB9",command=self.add_words2)
        b2= Button(frame2,text='新增',bg="#FFDAB9",command=self.enter_words)
        b1= Button(frame2,text='尋找',bg="#FFDAB9",command=self.dic)
        b4= Button(frame2,text='離開',bg="#33cc99",command=window.destroy)
        
        b1.pack(side='left')
        b2.pack(side='left')
        b3.pack(side='left')
        b4.pack(side='left')
        
        frame2.pack()
        

        self.text=Text(frame3)
        self.text.pack(side=LEFT,fill=Y)
        frame3.pack()
        
        window.mainloop()   
        
    def add_words2(self):
        with open('MyWordbank.csv','a',newline='',encoding='utf-8-sig') as wdFile:
            wdwriter=csv.writer(wdFile, dialect = 'excel')
            words=str(self.entry.get())
        
            wdwriter.writerow([words,self.current_wrds[0],self.current_wrds[1]])
        #print([nvoc,ndef,nex])
        
        isOK=tkinter.messagebox.showinfo(title='Success',message='Success to add!')
        print(isOK)    


    def dic(self):
        words=str(self.entry.get())
        with open('英文單字.csv',encoding='utf-8-sig') as csvfileobj:
            readerobj=csv.reader(csvfileobj)
            wrdsbook={}
            try:
                for row in readerobj:
                    wrdsbook[row[0].strip()]=[row[1],row[2]]
                self.current_wrds = wrdsbook[words]
                self.text.insert(END,self.current_wrds[0]+"\n"+self.current_wrds[1]+"\n")
            except:
                self.text.insert(END,"Not found.\n")
            

Main() 

```
